```python
import numpy as np
from random import randint
```

```python
def throw():
    return randint(0,5)+1
```

```python
def run(n):
    prize = np.array([0] + [1]*(n-1))
    #print(prize)
    output = []
    pos = 0
    turn = 0
    while any(prize):
        go = throw()
        pos += go
        turn += 1
        pos %= n
        if prize[pos] == 1:            
            prize[pos] = 0
            output.append(turn)
        #print(f'Throws {go} on turn {turn}. Current prize pot: {prize}')
    return turn, np.array(output)
```

```python
tot = 0
n = 24
tot_arr = np.array([0] * (n-1))
for i in range(200000):
    turn, output = run(n)
    tot += turn
    tot_arr += output
    if (i+1) % 10000 == 0:
        print(f'Avg {tot/(i+1)} on {i+1}th run\nAvg {tot_arr/(i+1)} on {i+1}th run')
```
